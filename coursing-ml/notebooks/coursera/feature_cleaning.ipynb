```python
def extract_category(x):
    x = str(x)
    if len(str(x).split(',')) > 1:
        return str(x).split(',')[1]
    else:
        return ""
```

```python
def extract_subcategory(x):
    x = str(x)
    if len(str(x).split(',')) > 1:
        return str(x).split(',')[2]
    else:
        return ""
```

```python
df['subcategory'] = df['category'].apply(extract_subcategory)
df['category'] = df['category'].apply(extract_category)
df
```

```python
def process_difficulty(x):
    x = str(x)
    x = x.split(' ')[0]
    if x == 'Professional':
        x = 'Advanced'
    if x not in ['Beginner', 'Intermediate', 'Advanced']:
        x = 'Beginner'
    return x
```

### Duration


```python
df.duration.value_counts()[:30]
```

```python
def extract_duration_hours(x):
    try:
        if len(str(x).split(' ')) > 3:
            if str(x).split(' ')[1] in ['hour', 'hours', 'hr']:
                return float(str(x).split(' ')[0])
            return float(str(x).split(' ')[1])
        elif len(str(x).split(' ')) == 2:
            return float(str(x).split(' ')[0])
        else:
            return 0
    except ValueError:
        return None
```

```python
df['hours_months'] = df['duration'].apply(extract_duration_hours)
df
```

### duration_week

```python
def extract_duration_hours(x):
    if len(str(x).split(' ')) > 3:
        return float(str(x).split(' ')[3])
    elif len(str(x).split(' ')) == 2:
        return float(str(x).split(' ')[0])
    else:
        return 0
```

```python
df['hours_week'] = df['duration_week'].apply(extract_duration_hours)
df
```

### n_ratings

```python
def convert_ratings(x):
    if len(str(x).split(' ')) > 1:
        return float(str(x).split(' ')[0].replace(",", ""))
    else:
        return 1
```
