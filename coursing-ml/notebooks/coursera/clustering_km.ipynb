```python
from feature_engine.imputation import ArbitraryNumberImputer

# set up the imputer
arbitrary_imputer = ArbitraryNumberImputer(arbitrary_number=0, variables=['rating'])

# fit the imputer
arbitrary_imputer.fit(df)

# transform the data
df= arbitrary_imputer.transform(df)
```

```python
from sklearn.cluster import KMeans

X = df
clusters = []

for i in range(1, 30):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 30)), y=clusters, ax=ax)
ax.set_title('Searching for Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')

plt.show()
```

```python
# 7 cluster
km3 = KMeans(n_clusters=8).fit(X)

df['Labels'] = km3.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(X['total_hours'], X['enrolled'], hue=X['Labels'])
plt.show()
```
