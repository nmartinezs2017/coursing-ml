```python
def process_text(x):
    x = str(x)
    x = x.replace("\n", "")
    x = x.replace("\r", "")
    x = x.replace(",", "")
    x = x.replace("\xa0", " ")
    return re.sub(",([\w]+)", r"\1", x)
```
