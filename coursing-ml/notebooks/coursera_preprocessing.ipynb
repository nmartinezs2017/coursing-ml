```python

# set up the imputer
imputer = CategoricalImputer(variables=['description'], fill_value='')

# fit the imputer
imputer.fit(df)

# transform the data
df = imputer.transform(df)
```

### Difficulty

```python
df.difficulty.value_counts()
```

```python
df[df["difficulty"].isnull()]
```

```python
mask_b = df['difficulty'] == 'Beginner Level'
mask_a = df['difficulty'] == 'Advanced Level'
mask_i = df['difficulty'] == 'Intermediate Level'

df['difficulty'] = df['difficulty'].where(mask_b | mask_a | mask_i, other='Advanced Level')
```
