# Categorical Variable Encoding

## Institution and Instructor: CountFrequency

```python

from feature_engine.encoding import CountFrequencyEncoder

# set up the encoder
encoder = CountFrequencyEncoder(encoding_method='frequency',
                         variables=['instructor', 'institution'])

# fit the encoder
encoder.fit(df)

# transform the data
df= encoder.transform(df)
```

### Difficulty: Ordinal

```python
df['difficulty'] = df['difficulty'].map({'Beginner':0, 'Intermediate':1, 'Advanced':2})
```

```python
df
```

## Category and subcategory and section: OneHotEncoder

```python
from feature_engine.encoding import OneHotEncoder
```

```python
# set up the encoder
encoder = OneHotEncoder(variables=['category','subcategory', 'section', 'language'], drop_last=False)

# fit the encoder
encoder.fit(df)

# transform the data
df= encoder.transform(df)
```

```python
df
```

### Rating: Variable Discretation

```python
df.rating.value_counts()
```

```python
import numpy as np
from feature_engine.discretisation import ArbitraryDiscretiser

user_dict = {'rating': [3.0, 4.1, 4.4, 4.5, 4.6, 4.7, 4.8, 4.9, np.Inf]}

transformer = ArbitraryDiscretiser(
    binning_dict=user_dict, return_object=False, return_boundaries=False)
df = transformer.fit_transform(df)

df
```
