```python
def extract_category(x):
    x = str(x)
    if len(str(x).split(',')) > 1:
        return str(x).split(',')[1]
    else:
        return ""
```

```python
def extract_subcategory(x):
    x = str(x)
    if len(str(x).split(',')) > 1:
        return str(x).split(',')[2]
    else:
        return ""
```

```python
df['subcategory'] = df['category'].apply(extract_subcategory)
df['category'] = df['category'].apply(extract_category)
df
```

### views

```python
def convert_views(x):
    if len(str(x).split(',')) > 2:
        new_number = str(x).split(',')[0] + str(x).split(',')[1]
        return float(new_number)
    else:
        return float(str(x).replace(",", ""))
```
